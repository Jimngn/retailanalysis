```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('retail_sales_dataset.csv', index_col = 'Transaction ID')
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df.describe()
plt.figure(figsize=(14,6))
plt.title("Complete Data")
sns.lineplot(data = df, dashes = False)
```

```python
plt.figure(figsize=(14,6))
plt.title("Complete Data")
sns.lineplot(data = df['Total Amount'], dashes = False)
```

```python
male = df.loc[df['Gender'] == 'Male'].count()
```

```python
gender = df.loc[df['Gender'].count()
```

```python
plt.figure(figsize = (10,6))
plt.title('Barchart girl vs boys')
sns.barplot(x = ('Male', 'Female'), y = (male, female))
```

```python
by_date = df.sort_values(['Date'], ascending = True)
total_sales_by_day = df.groupby('Date')['Total Amount'].sum()
total_sales_by_day
```

```python
plt.figure(figsize=(16,6))
plt.title('Total sales day by day')
sns.lineplot(data=total_sales_by_day)
plt.ylim(0)
```
